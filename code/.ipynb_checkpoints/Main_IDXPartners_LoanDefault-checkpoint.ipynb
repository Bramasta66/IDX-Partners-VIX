# Something(s) worth to investigate:

1. **funded_amnt vs funded_amnt_inv**
- The number of unique values in funded_amnt (1,354) and funded_amnt_inv (9,854) is significantly different, even though they should conceptually be close.
- funded_amnt_inv seems to have fractional values, possibly indicating partial funding by investors. This warrants checking for anomalies or understanding if all loans are partially funded.

2. **home_ownership**
The categories include NONE and ANY, which are unusual. Typically, homeownership options are RENT, OWN, or MORTGAGE. Verify if these values are data entry errors or if they hold specific meanings.

3. **loan_status**
There are statuses like 'Does not meet the credit policy. Status: Fully Paid' and 'Does not meet the credit policy. Status: Charged Off'. It would be useful to investigate why loans that didn't meet the policy were still issued and resulted in such statuses.

4. **emp_length**
- There are nan (missing) values, which may require imputation or separate analysis.
- Consider whether < 1 year and 1 year are treated consistently in downstream analysis.

5. **verification_status**
The distinct values seem fine, but further analysis can check how verification_status correlates with loan defaults.

6. **pymnt_plan**
Only two values ('y' and 'n') exist, but since 'y' is generally rare, it might be an anomaly worth checking.

7. **desc**
- Many nan values exist in the desc field, indicating borrowers may not always provide descriptions.
- Also, the presence of HTML tags (`<br>`) suggests the need for text cleaning.

8. **id vs member_id**
Both id and member_id are unique (466,285 values). Confirm whether both fields are necessary or if id is simply a primary key.

9. **emp_title**
With 205,476 distinct values, emp_title is highly inconsistent, possibly due to free-text input. We might want to apply text normalization or categorize the data.

10. **issue_d**
The dates are represented as 'Dec-11', 'Nov-11', etc. We need to convert them to a standardized datetime format for analysis.

# Actionables (for now)
1. Drop Columns with Zero Data: Remove columns like annual_inc_joint, dti_joint, and open_acc_6m, as they provide no value.

2. Handle Missing Values: Columns like emp_title, emp_length, and desc have significant missing data. Consider imputing, dropping, or filling them based on your analysis goals.

3. Convert Dates: Columns like issue_d, earliest_cr_line, and last_pymnt_d are in object format. Converting them to datetime will help with time-based analysis.

4. Explore Outliers: Fields like int_rate, revol_util, and dti may contain extreme values that could impact your analysis.

5. Encode Categorical Variables: Variables like grade, sub_grade, and home_ownership need proper encoding for machine learning models.

```python
# Cleaning
# 1. Dropping Columns with Excessive Missing Data (e.g., > 50% missing)
threshold = 0.5
missing_ratio = df.isnull().sum() / len(df)
drop_cols = missing_ratio[missing_ratio > threshold].index
print(f"\nDropping Columns with > {threshold*100}% missing values: {list(drop_cols)}")
df.drop(columns=drop_cols, inplace=True)

# 2. Filling Missing Values
# Numerical: Fill with Median
num_cols = df.select_dtypes(include=['float64', 'int64']).columns
df[num_cols] = df[num_cols].apply(lambda x: x.fillna(x.median()))

# Categorical: Fill with Mode
cat_cols = df.select_dtypes(include=['object']).columns
df[cat_cols] = df[cat_cols].apply(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else 'Unknown'))

# 3. Converting Data Types if Necessary (Example: Convert Dates)
for col in ['issue_d', 'earliest_cr_line', 'last_pymnt_d', 'next_pymnt_d', 'last_credit_pull_d']:
    df[col] = pd.to_datetime(df[col], format='%b-%y', errors='coerce')

# Final Check
print("\n--- Cleaned Data Info ---")
df.info()
```

# Label Encode, One-hot Encode, Conveting datetime features

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

# Assuming df_cleaned is your DataFrame

# Label Encoding for ordinal categorical features
label_encoder = LabelEncoder()
for col in ['grade', 'sub_grade']:  # Ordinal categories
    df[col] = label_encoder.fit_transform(df[col])

# One-Hot Encoding for nominal categorical features
nominal_cols = ['term', 'home_ownership', 'verification_status', 'loan_status', 'purpose']
df = pd.get_dummies(df, columns=nominal_cols)

# Converting datetime features to meaningful values
df['issue_d'] = pd.to_datetime(df['issue_d'])
df['earliest_cr_line'] = pd.to_datetime(df['earliest_cr_line'])
df['issue_year'] = df['issue_d'].dt.year
df['issue_month'] = df['issue_d'].dt.month
df['credit_history_length'] = (df['issue_d'] - df['earliest_cr_line']).dt.days / 365

# Drop original datetime columns
df = df.drop(columns=['issue_d', 'earliest_cr_line'])
```

# Modeling, not Final However it works .. so ..

```python
from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, roc_auc_score, confusion_matrix

# 1. Data Preparation
numerical_features = df_cleaned.select_dtypes(include=['int64', 'float64']).columns.tolist()
numerical_features.remove('target')  # Remove target from numerical features

categorical_features = ['addr_state', 'initial_list_status']

# Create a ColumnTransformer
transformer = ColumnTransformer(transformers=[
    ('num', StandardScaler(), numerical_features),
    ('cat', OneHotEncoder(sparse_output=False, handle_unknown='ignore'), categorical_features)
], remainder='passthrough')  # Include boolean columns

# Apply the transformer
transformed_data = transformer.fit_transform(df_cleaned.drop('target', axis=1))
transformed_df = pd.DataFrame(transformed_data, columns=transformer.get_feature_names_out())

# Split data
X_train, X_test, y_train, y_test = train_test_split(transformed_df, df_cleaned['target'], test_size=0.2, random_state=42)

# Models
models = {
    "Logistic Regression": LogisticRegression(solver='liblinear', random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42),
    "XGBoost": XGBClassifier(random_state=42)
}

# 2. Model Training, Evaluation, and Visualization
for model_name, model in models.items():
    print(f"\n--- {model_name} ---")

    # Train
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    # Classification Report
    print("Classification Report:")
    print(classification_report(y_test, y_pred))

    # AUC-ROC
    print("AUC-ROC:", roc_auc_score(y_test, model.predict_proba(X_test)[:, 1]))

    # Confusion Matrix
    conf_matrix = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', 
                xticklabels=['Non-Default', 'Default'], 
                yticklabels=['Non-Default', 'Default'])
    plt.title(f'Confusion Matrix - {model_name}')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()

    # Feature Importance
    if model_name == "Logistic Regression":
        feature_importance = pd.DataFrame({'Feature': transformed_df.columns, 'Importance': abs(model.coef_[0])})
        feature_importance = feature_importance.sort_values(by='Importance', ascending=False)

        print("\nLogistic Regression Feature Importance (Absolute Coefficients):")
        print(feature_importance.head(10))

        plt.figure(figsize=(10, 6))
        sns.barplot(x='Importance', y='Feature', data=feature_importance.head(10))
        plt.title('Top 10 Feature Importance - Logistic Regression (Absolute Coefficients)')
        plt.show()

    elif model_name != "Logistic Regression":
        feature_importance = model.feature_importances_
        feature_names = transformed_df.columns
        feature_importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': feature_importance})
        feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

        print("\nFeature Importance:")
        print(feature_importance_df.head(10))

        plt.figure(figsize=(10, 6))
        sns.barplot(x='Importance', y='Feature', data=feature_importance_df.head(10))
        plt.title(f'Top 10 Feature Importance - {model_name}')
        plt.show()

    # 3. Cross-Validation
    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
    cv_scores = cross_val_score(model, X_train, y_train, cv=cv, scoring='roc_auc')

    print("\nCross-Validation AUC-ROC Scores:", cv_scores)
    print("Mean Cross-Validation AUC-ROC:", cv_scores.mean())
    print("Standard Deviation of Cross-Validation AUC-ROC:", cv_scores.std())
```

```python
from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, roc_auc_score, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Data Preparation
numerical_features = df_cleaned.select_dtypes(include=['int64', 'float64']).columns.tolist()
numerical_features.remove('target')  # Remove target from numerical features

categorical_features = ['initial_list_status']

# Create a ColumnTransformer
transformer = ColumnTransformer(transformers=[
    ('num', StandardScaler(), numerical_features),
    ('cat', OneHotEncoder(sparse_output=False, handle_unknown='ignore'), categorical_features)
], remainder='passthrough')  # Include boolean columns

# Apply the transformer
transformed_data = transformer.fit_transform(df_cleaned.drop('target', axis=1))
transformed_df = pd.DataFrame(transformed_data, columns=transformer.get_feature_names_out())

# Split data
X_train, X_test, y_train, y_test = train_test_split(transformed_df, df_cleaned['target'], test_size=0.2, random_state=42)

# Models
models = {
    "Logistic Regression": LogisticRegression(solver='liblinear', random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42),
    "XGBoost": XGBClassifier(random_state=42)
}

# 2. Model Training, Evaluation, and Visualization
for model_name, model in models.items():
    print(f"\n--- {model_name} ---")

    # Train
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    # Classification Report
    print("Classification Report:")
    print(classification_report(y_test, y_pred))

    # AUC-ROC
    print("AUC-ROC:", roc_auc_score(y_test, model.predict_proba(X_test)[:, 1]))

    # Confusion Matrix
    conf_matrix = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', 
                xticklabels=['Non-Default', 'Default'], 
                yticklabels=['Non-Default', 'Default'])
    plt.title(f'Confusion Matrix - {model_name}')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()

    # Feature Importance
    if model_name == "Logistic Regression":
        feature_importance = pd.DataFrame({'Feature': transformed_df.columns, 'Importance': abs(model.coef_[0])})
        feature_importance = feature_importance.sort_values(by='Importance', ascending=False)

        print("\nLogistic Regression Feature Importance (Absolute Coefficients):")
        print(feature_importance.head(10))

        plt.figure(figsize=(10, 6))
        sns.barplot(x='Importance', y='Feature', data=feature_importance.head(10))
        plt.title('Top 10 Feature Importance - Logistic Regression (Absolute Coefficients)')
        plt.show()

    elif model_name != "Logistic Regression":
        feature_importance = model.feature_importances_
        feature_names = transformed_df.columns
        feature_importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': feature_importance})
        feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

        print("\nFeature Importance:")
        print(feature_importance_df.head(10))

        plt.figure(figsize=(10, 6))
        sns.barplot(x='Importance', y='Feature', data=feature_importance_df.head(10))
        plt.title(f'Top 10 Feature Importance - {model_name}')
        plt.show()

    # 3. Cross-Validation
    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
    cv_scores = cross_val_score(model, X_train, y_train, cv=cv, scoring='roc_auc')

    print("\nCross-Validation AUC-ROC Scores:", cv_scores)
    print("Mean Cross-Validation AUC-ROC:", cv_scores.mean())
    print("Standard Deviation of Cross-Validation AUC-ROC:", cv_scores.std())
```

# More Modeling Using SMOTE and Removing High Cardinality

```python
from imblearn.over_sampling import SMOTE

# Assuming your cleaned DataFrame is named df_cleaned

# 1. Data Preparation - High Cardinality
# Binning 'annual_inc'
bins_annual_inc = [0, 30000, 60000, 90000, 120000, float('inf')]
labels_annual_inc = ['0-30k', '30-60k', '60-90k', '90-120k', '120k+']
df_cleaned['annual_inc_binned'] = pd.cut(df_cleaned['annual_inc'], bins=bins_annual_inc, labels=labels_annual_inc)

# One-hot encode the binned column
df_cleaned = pd.get_dummies(df_cleaned, columns=['annual_inc_binned'], drop_first=True)

# Drop the original 'annual_inc' column
df_cleaned = df_cleaned.drop('annual_inc', axis=1)

# Drop the 'total_rec_prnc'
#df_cleaned = df_cleaned.drop('total_rec_prncp', axis=1)

# Drop the 'funded_amnt_inv'
#df_cleaned = df_cleaned.drop('funded_amnt_inv', axis=1)

# Drop the 'total_pymnt'
df_cleaned = df_cleaned('total_pymnt', axis=1)

# Drop the 'loan_amnt'
df_cleaned = df_cleaned('loan_amnt', axis=1)

# Binning 'credit_history_length'
bins_credit_history = [0, 5, 10, 15, 20, float('inf')]
labels_credit_history = ['0-5', '5-10', '10-15', '15-20', '20+']
df_cleaned['credit_history_length_binned'] = pd.cut(df_cleaned['credit_history_length'], bins=bins_credit_history, labels=labels_credit_history)

# One-hot encode the binned column
df_cleaned = pd.get_dummies(df_cleaned, columns=['credit_history_length_binned'], drop_first=True)

# Drop the original 'credit_history_length' column
df_cleaned = df_cleaned.drop('credit_history_length', axis=1)

# Drop 'addr_state' (for now)
df_cleaned = df_cleaned.drop('addr_state', axis=1, errors='ignore')

# Separate features and target
X = df_cleaned.drop('target', axis=1)
y = df_cleaned['target']

# Identify numerical and categorical features
numerical_features = X.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_features = X.select_dtypes(include=['object', 'bool']).columns.tolist()

# Create a ColumnTransformer
transformer = ColumnTransformer(transformers=[
    ('num', StandardScaler(), numerical_features),
    ('cat', OneHotEncoder(sparse_output=False, handle_unknown='ignore'), categorical_features)
], remainder='passthrough')

# Apply the transformer
transformed_data = transformer.fit_transform(X)
transformed_df = pd.DataFrame(transformed_data, columns=transformer.get_feature_names_out())

# Split data
X_train, X_test, y_train, y_test = train_test_split(transformed_df, y, test_size=0.2, random_state=42)

# 2. Apply SMOTE for Oversampling
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_train, y_train)

# Models
models = {
    "Logistic Regression": LogisticRegression(solver='liblinear', random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42),
    "XGBoost": XGBClassifier(random_state=42)
}

# 3. Model Training, Evaluation, and Visualization
for model_name, model in models.items():
    print(f"\n--- {model_name} ---")

    # Train (on resampled data)
    model.fit(X_resampled, y_resampled)
    y_pred = model.predict(X_test)

    # Classification Report
    print("Classification Report:")
    print(classification_report(y_test, y_pred))

    # AUC-ROC
    print("AUC-ROC:", roc_auc_score(y_test, model.predict_proba(X_test)[:, 1]))

    # Confusion Matrix
    conf_matrix = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', 
                xticklabels=['Non-Default', 'Default'], 
                yticklabels=['Non-Default', 'Default'])
    plt.title(f'Confusion Matrix - {model_name}')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()

    # Feature Importance
    if model_name == "Logistic Regression":
        feature_importance = pd.DataFrame({'Feature': transformed_df.columns, 'Importance': abs(model.coef_[0])})
        feature_importance = feature_importance.sort_values(by='Importance', ascending=False)

        print("\nLogistic Regression Feature Importance (Absolute Coefficients):")
        print(feature_importance.head(10))

        plt.figure(figsize=(10, 6))
        sns.barplot(x='Importance', y='Feature', data=feature_importance.head(10))
        plt.title('Top 10 Feature Importance - Logistic Regression (Absolute Coefficients)')
        plt.show()

    elif model_name != "Logistic Regression":
        feature_importance = model.feature_importances_
        feature_names = transformed_df.columns
        feature_importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': feature_importance})
        feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

        print("\nFeature Importance:")
        print(feature_importance_df.head(10))

        plt.figure(figsize=(10, 6))
        sns.barplot(x='Importance', y='Feature', data=feature_importance_df.head(10))
        plt.title(f'Top 10 Feature Importance - {model_name}')
        plt.show()

    # 4. Cross-Validation (on resampled training data)
    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
    cv_scores = cross_val_score(model, X_resampled, y_resampled, cv=cv, scoring='roc_auc')

    print("\nCross-Validation AUC-ROC Scores:", cv_scores)
    print("Mean Cross-Validation AUC-ROC:", cv_scores.mean())
    print("Standard Deviation of Cross-Validation AUC-ROC:", cv_scores.std())
```
